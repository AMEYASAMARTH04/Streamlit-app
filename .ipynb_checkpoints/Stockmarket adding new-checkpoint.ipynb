```python
import pandas as pd
import yfinance as yf

# Past buy records — suppose kal kharida
portfolio = pd.DataFrame({
    'Company': ['RELIANCE.NS', 'ICICIBANK.NS'],
    'Buy_Date': ['2025-07-12', '2025-07-12'],
    'Buy_Price': [2700, 950],
    'Qty': [100, 50]
})

# Get current prices
for idx, row in portfolio.iterrows():
    ticker = yf.Ticker(row['Company'])
    current = ticker.history(period="1d")['Close'].iloc[-1]
    portfolio.loc[idx, 'Current_Price'] = current
    portfolio.loc[idx, 'P&L%'] = round(((current - row['Buy_Price']) / row['Buy_Price']) * 100, 2)

portfolio.to_csv('portfolio_pnl.csv', index=False)
print(portfolio)
```

```python
from newspaper import Config, Source

user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'
config = Config()
config.browser_user_agent = user_agent

url = 'https://www.hindustantimes.com/business'
paper = Source(url, config=config)
paper.build()

print(f"Found {len(paper.articles)} articles")

for article in paper.articles:
    print(article.url)
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.hindustantimes.com/business'
headers = {
    'User-Agent': 'Mozilla/5.0'
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, 'html.parser')

# HT headlines ka example CSS selector (inspect karke adjust karna!)
for headline in soup.select('h3 a'):
    print(headline.text.strip())
    print(headline['href'])
```
